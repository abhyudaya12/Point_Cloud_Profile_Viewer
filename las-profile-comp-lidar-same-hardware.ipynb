```python
vertex_1 = gpd.read_file("/kaggle/input/las-data-shape-files/medium_profile_line_1_Result_Vertex_1.shp")

vertex_2 = gpd.read_file("/kaggle/input/las-data-shape-files/medium_profile_line_1_Result_Vertex_2.shp")
```

```python
#This code can also randomly remove any no. of points from the geodataframe so that the plotting could be faster.
#the advantage of that is it does not affect the shape of the plot.
num_toremove = 24007
indices_to_remove1 = vertex_1.sample(n=num_toremove).index
indices_to_remove2 = vertex_2.sample(n=num_toremove).index
vertex_1_1 = vertex_1.drop(indices_to_remove1)
vertex_2_1 = vertex_2.drop(indices_to_remove2)
print(vertex_1_1.shape[0])
print(vertex_2_1.shape[0])
print(type(vertex_1))
vertex_1_g=vertex_1_1.loc[:, ["geometry"]].copy()
vertex_2_g=vertex_2_1.loc[:, ["geometry"]].copy()
vertex_1_g.head()
```

Next Comes the 2D plot of the two vertices using Plotly where you can utilize the menu on top to manipulate the plot for better viewing and the scale of the plot will automatically change based on the viewpoint.

```python
# Extract x, y, and z coordinates
x1, y1, z1 = vertex_1_g.geometry.x, vertex_1_g.geometry.y, vertex_1_g.geometry.z
x2, y2, z2 = vertex_2_g.geometry.x, vertex_2_g.geometry.y, vertex_2_g.geometry.z

# Create traces for each set of points
trace1 = go.Scatter(
    x=y1,
    y=z1,
    mode='lines+markers',  # Set to 'lines' for a line connecting points
    marker=dict(size=5, color='red', line=dict(width=2)),  # Customize markers
    name='Vertex 1'
)

trace2 = go.Scatter(
    x=y2,
    y=z2,
    mode='lines+markers',  # Set to 'lines' for a line connecting points
    marker=dict(size=5, color='blue', line=dict(width=2)),  # Customize markers
    name='Vertex 2'
)

# Create the layout for the plot
layout = go.Layout(
    title='Plot of Points: 0216',
    xaxis=dict(title='X'),
    yaxis=dict(title='Y'),
    scene=dict(
        # Adjust camera position as needed
        camera=dict(
            up=dict(x=0, y=0),
            center=dict(x=0, y=0),
            eye=dict(x=1.2, y=1.2)
        )
    )
)

# Combine traces and layout into a figure
fig = go.Figure(data=[trace1, trace2], layout=layout)

# Display the in
fig.show()
```
