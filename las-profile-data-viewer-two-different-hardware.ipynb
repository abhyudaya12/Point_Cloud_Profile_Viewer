```python
vertex_1 = gpd.read_file("/kaggle/input/las-data-shape-files/profile_shape_Result_Vertex_1.shp")

vertex_2 = gpd.read_file("/kaggle/input/las-data-shape-files/profile_shape_Result_Vertex_2.shp")

line_strings = gpd.read_file("/kaggle/input/las-data-shape-files/profile_shape.shp")
```

```python
print(vertex_1.shape[0])
print(vertex_2.shape[0])
print(type(vertex_1))
vertex_1_g=vertex_1.loc[:, ["geometry"]].copy()
vertex_2_g=vertex_2.loc[:, ["geometry"]].copy()
vertex_1_g.head()
```

```python

def plot_3d(elev,azim):
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(vertex_1_g.geometry.x, vertex_1_g.geometry.y, vertex_1_g.geometry.z, c='red', s=10)
    ax.scatter(vertex_2_g.geometry.x, vertex_2_g.geometry.y, vertex_2_g.geometry.z, c='blue', s=10)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('3D Plot of Points')
    ax.view_init(elev=elev, azim=azim)
    plt.show()
```

```python
# Extract x, y, and z coordinates
x1, y1, z1 = vertex_1_g.geometry.x, vertex_1_g.geometry.y, vertex_1_g.geometry.z
x2, y2, z2 = vertex_2_g.geometry.x, vertex_2_g.geometry.y, vertex_2_g.geometry.z

# Create traces for each set of points
trace1 = go.Scatter3d(
    x=x1,
    y=y1,
    z=z1,
    mode='markers',  # Set to 'lines' for a line connecting points
    marker=dict(size=10, color='red', line=dict(width=2)),  # Customize markers
    name='Vertex 1'
)

trace2 = go.Scatter3d(
    x=x2,
    y=y2,
    z=z2,
    mode='markers',  # Set to 'lines' for a line connecting points
    marker=dict(size=10, color='blue', line=dict(width=2)),  # Customize markers
    name='Vertex 2'
)

# Create the layout for the plot
layout = go.Layout(
    title='3D Plot of Points',
    xaxis=dict(title='X'),
    yaxis=dict(title='Y'),
    scene=dict(
        # Adjust camera position as needed
        camera=dict(
            up=dict(x=0, y=0, z=1),
            center=dict(x=0, y=0, z=0),
            eye=dict(x=1.2, y=1.2, z=1.2)
        )
    )
)

# Combine traces and layout into a figure
fig = go.Figure(data=[trace1, trace2], layout=layout)

# Display the interactive plot
fig.show()
```
